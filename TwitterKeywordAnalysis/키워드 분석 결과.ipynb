```
import pandas as pd
```

```
twit_cacao = pd.read_csv('/gdrive/My Drive/카카오 키워드 분석/카카오 트위터 크롤링.csv')
```

```
text = ""
for val in twit_cacao['내용']:
    text += val
```

```
text_list = text.replace("\n", " ").replace("\\n", " ").replace(":", " ").split()
```

```
keyword = set(text_list)
```

```
keyword = list(keyword)
```

```
keyword[:5]
```

```
word_count = []

for word in keyword:
    count = 0
    for word2 in text_list:
        if word == word2:
            count += 1
    word_count.append([word, count])
```

```
word_count #[키워드, 횟수]
```

```
while True:
    veri = 0
    for i in range(len(word_count)-1):
        if word_count[i][1] < word_count[i+1][1]:
            veri += 1
            word_count[i], word_count[i+1] = word_count[i+1], word_count[i]
    if veri == 0:
        break
```
