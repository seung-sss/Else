## Main : 360도로 움직이는 시뮬레이션

각A와 빗변을 아는 경우
- 높이 = 빗변 * sinA
- 밑변 = 빗변 * cosA


```
import math
import numpy as np
from matplotlib import pyplot as plt

def make_simul(tMax):
    res = []
    for _ in range(tMax):
        angle = np.random.randint(360)
        res.append([math.cos(angle), math.sin(angle)])

    steps = np.array(res)
    positions = np.cumsum(steps, axis = 0)
    return positions
```

```
tMax = 200
nStories = 1000
```

```
for _ in range(10):
    _tmp = make_simul(tMax)
    plt.plot(_tmp[:,0], _tmp[:,1])
```
