```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
data =pd.read_csv('/gdrive/My Drive/전산통계/원본 자료/data/wisc_bc_data.csv', sep=',')
data.head()
```

```
var = data.columns[2:] # id, diagnosis 제외한 변수들 리스트로 저장
var
```

```
data['diagnosis'].unique() # B: Benign(양성), M: Malignant(악성)
```

```
from scipy import stats
```

```
pval = dict() 

for value in var: # loop 돌며 각 변수에 대해 t-test진행
    temp_B = data[data['diagnosis'] == 'B'][value]
    temp_M = data[data['diagnosis'] == 'M'][value]
    res = stats.ttest_ind(temp_B, temp_M)
    pval[value] = res[1] # 변수명을 key로, p-value를 value로 dict 만듦 
```

```
pval_sorted = sorted(pval.items(), key = lambda x:x[1]) # p-value 기준 오름차순으로 정렬
pval_sorted[:5] # p-value값이 가장 작은 5가지 변수
```
