```
import random
random.randint(0, 1)
```

```
def hit_or_miss(x, y, img): #img는 2차원의 그림 / img[row = y, column = x]
    if (0 in img[y,:x]) and (0 in img[y,x:]) and (0 in img[:y,x]) and (0 in img[y:,x]):
        return True
    else:
        return False
```

```
def calc_area(count, img): # 시뮬레이션을 통해 넓이 구하기
    cnt = 0
    for _ in range(count):
        tempX = random.randint(0, 423)
        tempY = random.randint(0, 472)
        if hit_or_miss(tempX, tempY, img) == True:
            cnt += 1
    a, b = img.shape
    return round((a*b) * (cnt/count), 4)
```

```
print(img.shape[0] * img.shape[1]) # 그림 전체 면적
print(calc_area(1000, img)) # 해당 부분의 면적
```
