## 모델 정확도 검증

```
from sklearn import metrics
metric19 = metrics.confusion_matrix(y_test, pred)
print(metric19)
```

```
import numpy as np
np.sum(metric19)
```

```
accuracy = (metric19[0,0] + metric19[1,1]) / np.sum(metric19)
error_rate = 1 - accuracy
print(accuracy, error_rate)
```

## 최적 n사이즈 찾기

```
def accuracy(metric):
    import numpy as np

    length = len(metric)
    correct = 0
    for i in range(length):
        correct += metric[i, i]
    
    return correct / np.sum(metric) 
```

```
best_n = None
best_acc = 0

for i in range(1,100,2):
    knn = neighbors.KNeighborsClassifier(n_neighbors = i)
    knn.fit(Z_train, y_train)
    pred = knn.predict(Z_test)
    metric = metrics.confusion_matrix(y_test, pred)
    _tmp = accuracy(metric)
    if _tmp > best_acc:
        best_acc = _tmp
        best_n = i

print("(n=%s)일 때, 최적의 Accuracy(%0.4f)와 Error Rate(%0.4f) 갖는다!" % (best_n, best_acc, 1-best_acc))
```
