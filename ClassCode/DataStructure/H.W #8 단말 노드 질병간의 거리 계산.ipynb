* 리스트를 이용한 이진트리의 ADT
    * 완전 이진트리는 순차구조이므로 배열 형태로 구현할 수 있다. 단, 루트의 인덱스는 1로 시작한다.
      * t = [None]
    * append method: 마지막 위치에서 item을 삽입한다.
    * getChild method: item 을 찾고 이 위치 인덱스 k에 대해 t[2k], t[2k+1]이 좌, 우 자식값을 리턴한다. 
    * getParent method: item을 찾고 해당 인덱스 k에 대해 k//2 위치값이 부모값이다. 

```
# tree 만들기
class BinaryTree:

    def __init__(self):
        self.t = [None]

    def append(self, item):
        self.t.append(item)

    def size(self):
        return len(self.t) - 1
    
    def getChild(self, item):
        if item in self.t:
            k = self.t.index(item)
            lidx = 2 * k
            ridx = (2 * k) + 1
            if lidx <= self.size():
                lnode = self.t[lidx]
            else:
                lnode = None
            if ridx <= self.size():
                rnode = self.t[ridx]
            else:
                rnode = None
            return lnode, rnode
        else:
            print("item not found!")

    def getParent(self, item):
        if item in self.t:
            k = self.t.index(item)
            pidx = k//2
            if pidx > 0:
                return self.t[pidx]
            else:
                return None
        else:
            print('item not found~')

tree = BinaryTree()
for i in range(12):
    tree.append(chr(65+i))
```

```
def getDist(item1, item2):    
    d1 = disease.getParent(item1)
    d2 = disease.getParent(item2)
    count = 1

    while d1 != d2:
        d1 = disease.getParent(d1)
        d2 = disease.getParent(d2)
        count += 1

    return "%s/3" %count
```
