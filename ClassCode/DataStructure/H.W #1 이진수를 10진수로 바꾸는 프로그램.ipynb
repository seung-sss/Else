```
# 이진수를 십진수로 바꾸는 함수

def bin_to_dec(bin):
    bin = str(bin)
    bin_list = list(bin)
    bin_list.reverse()

    sum = 0
    for i in range(len(bin_list)):
        sum += int(bin_list[i])*(2**i)
        
    return sum
```

```
# 실행

a = bin_to_dec(101011111000)
print(a)
```

```
# 십진수를 이진수로 바꾸기(확인용)

dec = 2808
bin=[]
while True:
    fraction = dec % 2
    bin.append(fraction)
    dec=dec//2
    if dec < 1: break

bin.reverse()
print (bin)
```
