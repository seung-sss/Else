```
import numpy as np

f = open("/gdrive/My Drive/알고리즘/강의자료(원본)/data/colorPaper.txt", 'r')
n = int(f.readline())
a = []
for i in range(n):
    a.append(list(map(int,f.readline().split())))

def divide(a, nWhite = 0, nBlue = 0):
    n = a.shape[0]
    if np.sum(a) == n**2:
        return nWhite, nBlue + 1
    elif np.sum(a) == 0:
        return nWhite + 1, nBlue
    nWhite, nBlue = divide(a[:n//2, :n//2], nWhite, nBlue)
    nWhite, nBlue = divide(a[:n//2, n//2:], nWhite, nBlue)
    nWhite, nBlue = divide(a[n//2:, :n//2], nWhite, nBlue)
    nWhite, nBlue = divide(a[n//2:, n//2:], nWhite, nBlue)
    return nWhite, nBlue

a = np.array(a)
print(divide(a)[0])
print(divide(a)[1])
```
