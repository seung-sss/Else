```
class Sort:

    def __init__(self):
        pass

    def _exchange(self, x, i, j):
        x[i], x[j] = x[j], x[i]

    def _findMin(self, x, curIdx):
        minIdx = curIdx
        xMin = x[curIdx]
        for i in range(curIdx, len(x)):
            if x[i] < xMin:
                xMin = x[i]
                minIdx = i
        return minIdx

    def selectionSort(self, x):
        for i in range(len(x)):
            self._exchange(x, i, self._findMin(x, i))
        return x

    def bubbleSort(self, x):
        while True:
            cnt = 0
            for i in range(len(x) - 1):
                if x[i] > x[i+1]:
                    self._exchange(x, i, i+1)
                    cnt += 1
            if cnt == 0:
                break
        return x

    def insertSort(self, x):
        for curIdx in range(1, len(x)):
            for i in range(curIdx, 0, -1):
                if x[i] < x[i-1]:
                    self._exchange(x, i, i-1)
        return x

    def shellSort(self, x):
        h = len(x) // 2
        while h >= 1:
            for i in range(h):
                _tmp = x[i:len(x):h]
                self.insertSort(_tmp)

                incr = 0
                for j in range(len(_tmp)):
                    x[i+incr] = _tmp[j]
                    incr += h
            h = h //2
        return x

    def _merge(self, a, b):
        a.sort()
        b.sort()

        res = []

        while len(a) > 0 and len(b) > 0:
            if a[0] < b[0]:
                res.append(a.pop(0))
            else:
                res.append(b.pop(0))

        if len(a) == 0:
            res.extend(b)
        else:
            res.extend(a)
        return res

    def mergeSort(self, x):
        if len(x) <= 1:
            return x
        mid = len(x) // 2
        leftX = x[:mid]
        rightX = x[mid:]
        leftX = self.mergeSort(leftX)
        rightX = self.mergeSort(rightX)
        return self._merge(leftX, rightX)

x = [10, 3, 2, 6, 2, 1]
test = Sort()
# test.selectionSort(x)
# test.bubbleSort(x)
# test.insertSort(x)
# test.shellSort(x)
# test._merge([32,1,5], [3,1,8])
test.mergeSort(x)
```
