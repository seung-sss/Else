```
# ex4_4
f = open("/gdrive/My Drive/알고리즘/강의자료(원본)/data/ex4_4.txt", 'r')
n, c = map(int, f.readline().split())
a = []
for i in range(n):
    a.append(int(f.readline()))
a.sort()

def simul(n, c, a, mid):
    pivot = 0
    horse = 1
    for i in range(1, n):
        if a[i] - a[pivot] >= mid:
            horse += 1
            pivot = i
    return horse == c

mini = min(a)
maxi = max(a)
mid = (mini + maxi) // 2

for i in range(10):
    # print(mini, mid, maxi, simul(n, c, a, mid))
    if simul(n, c, a, mid) == False:
        maxi = mid
    else:
        mini = mid
        if maxi - mini == 1:
            break
    mid = (mini + maxi) // 2

print(mid)
```

```
# ex4_5
f = open("/gdrive/My Drive/알고리즘/강의자료(원본)/data/ex4_5.txt", 'r')
n = int(f.readline())
x = []
for i in range(n):
    x.append(list(map(int, f.readline().split())))

x_start = sorted(x, key = lambda x: x[0])
x_end = sorted(x, key = lambda x: x[1])

# 시작시간 빠른 순서대로 뽑기
max_meeting_start = 0
start = 0
for meet in x_start:
    if meet[0] >= start:
        start = meet[1]
        max_meeting_start += 1

# 종료시간 빠른 순서대로 뽑기        
stand = x_end[0][1]
res = [x_end[0]]
for i in range(1, n):
    if x_end[i][0] >= stand:
        res.append(x_end[i])
        stand = x_end[i][1]

print(max(max_meeting_start, len(res)))
```

```
# ex4_6
n = 5
x = [[172, 67], [183, 65], [180, 70], [170, 72], [181, 60]]

del_count = 0
for i in range(n):
    for j in range(n):
        if x[i][0] < x[j][0] and x[i][1] < x[j][1]:
            del_count += 1

print(n-del_count)
```

```
# ex 4_8
n, m = 5, 140
weight = [90, 50, 70, 100, 60]

weight.sort()
boat = 0

while True:
    if len(weight) == 1:
        weight.pop(0)
        boat += 1
        break
    else:
        candidate = []
        _tmp = weight.pop(0)
        for people in weight:
            if _tmp + people <= m:
                candidate.append(people)
        weight.pop(weight.index(max(candidate)))
        boat += 1
        # print(weight)

print(boat)
```

```
# ex4_9
n = 5
number = [2, 4, 5, 1, 3]

# n = 10
# number = [3, 2, 10, 1, 5, 4, 7, 8, 9, 6]

increase_list = []
increase_LR = ""

if number[0] < number[-1]:
    enter_num = number.pop(0)
    increase_LR += 'L'
else:
    enter_num = number.pop(-1)
    increase_LR += 'R'

while True:
    increase_list.append(enter_num)
    left = number[0]
    right = number[-1]
    if enter_num > left and enter_num > right:
        break
    elif enter_num > left and enter_num < right:
        enter_num = number.pop(-1)
        increase_LR += 'R'
    elif enter_num < left and enter_num > right:
        enter_num = number.pop(0)
        increase_LR += 'L'
    else:
        if left < right:
            enter_num = number.pop(0)
            increase_LR += 'L'
        else:
            enter_num = number.pop(-1)
            increase_LR += 'R'

print(len(increase_list))
print(increase_LR)
```

```
# ex4_10
n = 8
number = [5, 3, 4, 0, 2, 1, 1, 0]

def inversion(x, n):
    res = []
    for i in range(1, n+1):
        res.append(i)
    
    cnt = 0
    test = 0
    while True:
        for num in x:
            if res[cnt] < res[cnt+1]:
                res[cnt], res[cnt+1] = res[cnt+1], res[cnt]
                test += 1
                cnt += 1
                if test == num:
                    continue

inversion(number, n)

# 이 문제는 못풀겠습니다 ㅠㅠ
```
