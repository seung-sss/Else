```
def BFS(graph, start):
    queue = []
    queue += graph[start]
    visit = [start]

    while queue:
        item = queue.pop(0)
        if not item in visit:
            queue += graph[item]
            visit.append(item)
    return visit
```

```
# 서울 - 세종 사이 화재 = 서로 연결 끊기

korea = {'세종': set(['대구', '광주']),
         '서울': set(['평양', '인천', '강릉']),
         '강릉': set(['독도', '서울']),
         '광주': set(['세종', '여수']),
         '대구': set(['세종', '울산']),
         '평양': set(['서울']),
         '인천': set(['서울']),
         '독도': set(['강릉', '울산']),
         '여수': set(['광주', '부산']),
         '울산': set(['대구', '부산', '독도']),
         '부산': set(['여수', '울산']),
         }

all_city = set(korea.keys())

search = set(BFS(korea, '부산'))
# print(all_city)
# print(search)

if len(all_city - search) == 0:
    print("문제가 없습니다.")
else:
    _tmp = all_city - search
    print("%s가 연결되지 않습니다." %_tmp)
```
