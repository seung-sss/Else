```
# ex2_6
f = open("/gdrive/My Drive/알고리즘/강의자료(원본)/data/ex2_6.txt", 'r')
n = int(f.readline())
a = list(map(int, f.readline().split()))

def digit_sum(x):
    x = str(x)
    _tmp = []
    for char in x:
        _tmp.append(char)
    _tmp = list(map(int, _tmp))
    return sum(_tmp)

# digit_sum(123)

maxi = 0
for i in range(n):
    _tmp = digit_sum(a[i])
    if _tmp > maxi:
        maxi = _tmp
        maxi_idx = i

print(a[maxi_idx])   
```

```
# ex2_7(v.1)
n = 20
test = []
cnt = 0
for i in range(2, n+1):
    _tmp = []
    for j in range(2, i):
        _tmp.append(i % j)
    if 0 not in _tmp:
        cnt += 1
        test.append(i)

print(test)
print(cnt)
```

```
# ex2_7(v.2)
n = 20

def prime_number(x):
    if x != 1:
        for i in range(2, x):
            if x % i == 0:
                return False
    else:
        return False
    return True

cnt = 0
for i in range(1, n+1):
    if prime_number(i) == True:
        cnt += 1

print(cnt)
```

```
# ex2_8
f = open("/gdrive/My Drive/알고리즘/강의자료(원본)/data/ex2_8.txt", 'r')
n = int(f.readline())
a = list(f.readline().split())

def reverse(x):
    return x[::-1]

def isPrime(x):
    num = 2
    while num < x:
        if x % num == 0:
            return False
        num += 1
    return True

for num in a:
    _tmp = int(reverse(num))
    if isPrime(_tmp) == True:
        print(_tmp, end = " ")
```

```
# ex2_9
f = open("/gdrive/My Drive/알고리즘/강의자료(원본)/data/ex2_9.txt", 'r')
n = int(f.readline())
a = []
for i in range(n):
    _tmp = list(map(int, f.readline().split()))
    a.append(_tmp)

def change_dict(x): # x = list
    res = dict()
    for num in x:
        if num in res.keys():
            res[num] += 1
        else:
            res[num] = 1
    return res

total = []
for _a in a:
    a_dict = change_dict(_a)
    if len(a_dict) == 1:
        pick_dice = list(a_dict.keys())[0]
        money = 10000 + pick_dice * 1000
    elif len(a_dict) == 2:
        dice_1 = list(a_dict.keys())[0]
        dice_2 = list(a_dict.keys())[1]
        if dice_1 > dice_2:
            pick_dice = dice_1
        else:
            pick_dice = dice_2
        money = 1000 + pick_dice * 100
    elif len(a_dict) == 3:
        pick_dice = max(list(a_dict.keys()))
        money = pick_dice * 100
    total.append(money)

print(max(total))
```
