```
def bfs(graph, start, end):
    queue = []
    temps = []
    queue += graph[start]
    temps += [start] * len(graph[start])
    visit = [start]
    parents = ['None']

    while queue:
        item = queue.pop(0)
        temp = temps.pop(0)
        if not item in visit:
            queue += graph[item]
            temps += [item] * len(graph[item])
            visit.append(item)
            parents.append(temp)
            if item == end:
                visits = dict(zip(visit, parents))
                return visits
    return 0
```

```
import csv

f = open("/gdrive/My Drive/알고리즘/강의자료(원본)/data/subway.csv", 'r', encoding = 'utf-8-sig')
csvfile = csv.reader(f)

nodes = set()

for line in csvfile: #key 뽑아내기
    temp1 = line[0]
    temp2 = line[1]
    nodes.add(temp1)
    nodes.add(temp2)

f.close()
# nodes
```

```
f = open("/gdrive/My Drive/알고리즘/강의자료(원본)/data/subway.csv", 'r', encoding = 'utf-8-sig')
csvfile = csv.reader(f)

graph = {}
for node in nodes:
    graph[node] = set()

for line in csvfile:
    graph[line[0]].add(line[1])
    graph[line[1]].add(line[0])

f.close()
# graph
```

```
path = bfs(graph, '서울역(1)', '선릉(2)')
# path 

end = '선릉(2)'
minPath = [end]

while True:
    parent = path[end]
    minPath.append(parent)
    end = parent
    if parent == 'None': break

print(minPath[::-1][1:])
```
